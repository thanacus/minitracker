```python
import cv2
vidcap = cv2.VideoCapture('G:/Google Drive/ml/cornertracking/WIN_20230926_09_28_18_Pro.mp4')
success,image = vidcap.read()
count = 0
while success:
  cv2.imwrite(f"G:/Google Drive/ml/cornertracking/training-data/video1/obj_train_data/frame_%06d.PNG" % count, image)     # save frame as PNG file      
  success,image = vidcap.read()
  print('Read a new frame: ', success)
  count += 1
```
